```python
# mid-point circle drawing algorithm

from graphics import *
import time
def drawcircle(x0,y0,r):
    xi=0
    yi=r
    xtemp=[]
    ytemp=[]
    p=3-(2*r)
    while(xi<=yi):
        print("(",xi,",",yi,")")
        xtemp.append(xi)
        xtemp.append(yi)
        ytemp.append(yi)
        ytemp.append(xi)
        xnext=xi+1
        if(p<0):
            ynext=yi
            p=p+(4*xnext)+6
        else:
            ynext=yi-1
            p=p+(4*(xnext-ynext))+10
        xi=xnext
        yi=ynext
    x1=xtemp
    y1=ytemp
    x=[]
    y=[]
    for i in xtemp:
        x.append(i)
        x.append(i)
        x.append(-i)
        x.append(-i)
    for i in ytemp:
        y.append(i)
        y.append(-i)
        y.append(-i)
        y.append(i)    
    win = GraphWin('Ques2:- Mid Point Circle Drawing Algorithm by CSC/20/22', 600, 480)
    for i in range(len(x)):
        pt = Point(x[i]+x0,y[i]+y0)
        pt.draw(win) 
    time.sleep(10)
    win.close()
    
X0=int(input("Enter X cordinates of Center of Circle : "))
Y0=int(input("Enter Y cordinates of Center of Circle : "))
R=int(input("Enter Radius of Circle : "))
drawcircle(X0,Y0,R)
```
